```python
# importação de biblioteca
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
```

```python
# criação de um corpus lendo textos do disco
corpus = PlaintextCorpusReader('/content/drive/MyDrive/Arquivos','.*', encoding="ISO-8859-1")
```

```python
# LEITURA DOS ARQUIVO DO DISCO, PERCORRE OS REGISTROS E MOSTRAR O NOME DOS PRIMEIROS 100 ARQUIVOS
arquivos = corpus.fileids()
# primeiro arquivo
arquivos[1]
```

```python
# zero a 10
arquivos[0:10]
```

```python
# imprime todos os nomes
for a in arquivos:
  print(a)
```

```python
# acesso ao texto  do primeiro arquivo
texto = corpus.raw('1.txt')
texto
```

```python
# acesso ao texto as palavras de todos os arquivos do corpus
todo_texto =corpus.raw()
#todo_texto
```

```python
# obtençao de todas as palavras do corpus e  visualização da quantidade
palavras = corpus.words()
#acessando pelo indice
palavras[170]
```

```python
# quantidade
len(palavras)
```

```python
# usando o nltk, obtemos os stop word em inglês
stops = stopwords.words('english')
#stops = stopwords.words('portuguese')
stops
```

```python

# definição das cores que serão utilizados na nuvem de palavras
mapa_cores = ListedColormap(['orange','green','red','magenta'])
# criação de nuvem de palavras , com no maximo 100 palavras e utilizando as stops words
nuvem = WordCloud(background_color='white',
                  colormap= mapa_cores,
                  stopwords = stops,
                  max_words = 100)

# criação e visualização  da nuvem de palavras
nuvem.generate(todo_texto)
plt.imshow(nuvem)
```

```python
# criação de nova lista de palavras, removendo os stops
palavras_semstops = [p for p in palavras if p not in stops]
len(palavras_semstops)
```

```python
# remoção de pontuação , gerando uma lista sem stops words e sem pontuação
palavras_sem_pontuação = [p for p in palavras_semstops if p not in string.punctuation]
len(palavras_sem_pontuação)
```

```python
# calculo de frequência das palavras e visualização das mais comun
frequência = nltk.FreqDist(palavras_sem_pontuação)
frequência
```

```python
# mais comun
mais_comuns = frequência.most_common(100)
mais_comuns
```
